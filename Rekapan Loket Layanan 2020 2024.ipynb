**Range**

Range adalah rentang seluruh data. Selisih data terkecil (min) dengan data terbesar (max)

```python
#Membuat kolom range
descriptive_stat_awal.loc['Range'] = descriptive_stat_awal.loc['max'] - descriptive_stat_awal.loc['min']
```

Standar Deviasi

Standar deviasi adalah parameter statistik yang berguna untuk mengetahui seberapa jauh jarak antar satu data dengan data lain

```python
#Membuat standar deviasi
descriptive_stat_awal.loc['Variance'] = descriptive_stat_awal.loc['std']**2
```

Rentang Interkuartil (Interquartile Range), Q3-Q1

Berguna sebagai variabel untuk menghilangkan Outlier

```python
descriptive_stat_awal.loc['IQR'] = descriptive_stat_awal.loc['75%'] - descriptive_stat_awal.loc['25%']
```

Lower Bound = Q1 - 1.5 x IQR

```python
descriptive_stat_awal.loc['Lower Bound'] = descriptive_stat_awal.loc['25%'] - (1.5 * descriptive_stat_awal.loc['IQR'])
```

Upper Bound = Q3 + 1.5 x IQR

```python
descriptive_stat_awal.loc['Upper Bound'] = descriptive_stat_awal.loc['75%'] + (1.5 * descriptive_stat_awal.loc['IQR'])
```

Tujuan dari visualisasi data barisan waktu adalah untuk mengetahui tren dari pengajuan izin per bulan berdasarkan jenis bidang perizinan

```python
#Mengganti format bulan dari teks menjadi angka
encoded_loket_pelayanan1['BULAN'] = encoded_loket_pelayanan1['BULAN'].replace({'JANUARI': 1, 'FEBRUARI': 2, 'MARET': 3, 'APRIL': 4, 'MEI':5, 'JUNI':6, 'JULI':7, 'AGUSTUS':8, 'SEPTEMBER':9, 'OKTOBER':10, 'NOVEMBER':11, 'DESEMBER':12})
encoded_loket_pelayanan1['TAHUN'] = encoded_loket_pelayanan1['TAHUN'].astype(int)
#Membuat kolom tanggal dengan kolom bulan dan tahun
encoded_loket_pelayanan1['Date'] = pd.to_datetime(encoded_loket_pelayanan1['TAHUN'].astype(str) + '-' + encoded_loket_pelayanan1['BULAN'].astype(str) + '-01', format='%Y-%m-%d')
```

```python
#Agregasi data. Menjumlahkan data yang memiliki bidang dan tanggal yang sama
time_series = encoded_loket_pelayanan1.groupby(['LOKET LAYANAN','Date']).agg(
    PROSES = ('PROSES', 'sum'),
    KONSUL = ('KONSUL', 'sum'),
    TOLAK = ('TOLAK', 'sum'),
    AMBIL = ('AMBIL', 'sum'),
    JUMLAH = ('JUMLAH', 'sum')
)
```

```python
time_series.reset_index(inplace = True)
```

```python
#Mengurutkan berdasarkan tanggal dan bidang
time_series.sort_values(by=['Date','LOKET LAYANAN'], inplace = True)
```

```python
from matplotlib.dates import MonthLocator
#Visualisasi tren pengajuan izin dari jan 2020-jul 2024
plt.figure(figsize=(15,6))
sns.lineplot(x='Date', y='JUMLAH', data=time_series)
plt.xticks(rotation=25, ha='right', fontsize = 10)
plt.title('Time Series Plot Seluruh Lantai (Jan 2020 - Jul 2024)')
# Set month locators for x-axis
ax = plt.gca()
ax.xaxis.set_major_locator(MonthLocator(bymonth=[ 2,  4,  6,  8,  10,  12]))

plt.show()
```
